#### Typical time spent per week on social media?

```python
def calculate_mean(numbers):

    if not numbers:
        return None

    total = 0
    count = 0

    for num in numbers:
        total += num
        count += 1

    mean = total / count
    return mean
```

```python
calculate_mean(hrs_per_week)
```

Median time spent by students per week

```python
def calculate_median(data):
    # Sort the data first
    data.sort()

    n = len(data)

    if n % 2 != 0:
        return data[n // 2]
    else:
        mid1 = data[(n // 2) - 1]
        mid2 = data[n // 2]
        return (mid1 + mid2) / 2
```

```python
calculate_median(hrs_per_week)
```

Most popular social media platform

```python
def calculate_mode(data):
    frequency = {}

    for item in data:
        if item in frequency:
            frequency[item] += 1
        else:
            frequency[item] = 1

    max_freq = max(frequency.values())

    mode = [key for key, value in frequency.items() if value == max_freq]

    return mode
```

```python
calculate_mode(fav_platform)
```
